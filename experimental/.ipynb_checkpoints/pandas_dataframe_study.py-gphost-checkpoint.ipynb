```python
#plot with different marker, size, and color
import matplotlib.pyplot as plt
import pandas as pd
x = {'speed': [10, 15, 20, 18, 19],
     'meters' : [122, 150, 190, 230, 300],
     'type': ['phone', 'phone', 'gps', 'gps', 'car'],
     'weight': [0.2, 0.3, 0.1, 0.85, 0.3],
     'old': [1, 2, 4, 5, 8]}

m = pd.DataFrame(x)
print(m)

mkr_dict = {'gps': 'x', 'phone': '+', 'car': 'o'}
for kind in mkr_dict:
    d = m[m.type==kind]
    plt.scatter(d.meters, d.speed, 
                s = 100* d.weight, 
                c = d.old, 
                marker = mkr_dict[kind])
plt.show()

colors = ['r', 'b', 'k', 'g', 'm']
plt.scatter(m.meters, m.speed, s=30*m.weight, vmin=0, vmax=10, c=colors)
plt.show()


cm = plt.cm.get_cmap('hot')  # or your colormap of choice
plt.scatter(m.meters, m.speed, s=30*m.weight, c=m.old, cmap=cm)
plt.show()


markers = ['^', 'o', 'v', 's', 'd']
for px, py, c, s, t in zip(m.meters, m.speed, m.old, m.weight, markers):
    plt.scatter(px, py, marker=t, c=cm(c/10.), vmin=0, vmax=10, s=400*s+100)
plt.show()
```

```python

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Set some Pandas options
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 25)
variables = pd.DataFrame({'normal': np.random.normal(size=100), 
                       'gamma': np.random.gamma(1, size=100), 
                       'poisson': np.random.poisson(size=100)})
variables.cumsum(0).plot()
plt.show()

variables.cumsum(0).plot(subplots=True)
plt.show()

variables.cumsum(0).plot(secondary_y='normal')
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
plt.show()
```

```python
# pandas dataframe scatter plot
# merge two data frame, inner, outer, right, or left join
df1 = pd.DataFrame(dict(id=range(4), age=np.random.randint(18, 31, size=4)))
df2 = pd.DataFrame(dict(id=[0,0,1,1,2,2,4], score=np.random.random(size=7)))

print(df1)

print(df2)

print(pd.merge(df1, df2)) # default inner joining

print(pd.merge(df1, df2, how='outer'))

print(pd.merge(df1, df2, how='right'))

print(pd.merge(df1, df2, how='left'))

merged = pd.merge(df1, df2)
merged.plot(kind='scatter', x='age', y='score', c=merged['age'],s=merged['score']*500)

```

```python
# pandas dataframe concatenation and merging, binding / stacking
mb1 = pd.DataFrame(dict(id=range(4), age=np.random.randint(18, 31, size=4), score=[100, 200, 300, 400]))
mb2 = pd.DataFrame(dict(id=[0,10,11,13,2,3,4], score=np.random.random(size=7)))
mb1.set_index('id', inplace=True)
mb2.set_index('id', inplace=True)
print(mb1.shape, mb2.shape)
print(mb1)
print(mb2)

# concatenate rows
conc = pd.concat([mb1, mb2], axis=0)
print(conc)

#concatenate columns, binding
conc = pd.concat([mb1, mb2], axis=1)
print(conc)

# only data in both frame mb1, mb2
conc = pd.concat([mb1, mb2], axis=1, join='inner')
print(conc)

# If we wanted to use the second table to fill values absent from the first table, we could use combine_first.
conc = mb1.combine_first(mb2)
print(conc)

# add hierachy index
pd.concat([mb1, mb2], keys=['file1', 'file2']).head(100)

pd.concat([mb1, mb2], keys=['file1', 'file2']).index.is_unique
```

```python
# long and wide format of data
# stack, unstack, melt
cdystonia = pd.read_csv("/home/szong/projects/development/study/statistical-analysis-python-tutorial-master/data/cdystonia.csv", index_col=None)
print(cdystonia.head(20))
```

```python
# The stack method rotates the data frame so that columns are represented in rows:
stacked = cdystonia.stack()
print(stacked.head(20))
```

```python
# To complement this, unstack pivots from rows back to columns.
print(stacked.unstack().head(20))   
```

```python
# For this dataset, it makes sense to create a hierarchical index based on the patient and observation:
cdystonia2 = cdystonia.set_index(['patient','obs'])
print(cdystonia2.head(20))
```

```python
# If we want to transform this data so that repeated measurements are in columns, we can unstack the twstrs measurements according to obs.
twstrs_wide = cdystonia2['twstrs'].unstack('obs')
print(twstrs_wide.head())
```

```python
# unstack to wide format
cdystonia_wide = cdystonia.set_index(['patient','site','id','treat','age','sex','week'])['twstrs'].unstack('week')
print(cdystonia_wide.head(20))
```

```python
cdystonia_wide.reset_index(inplace=True)
print(cdystonia_wide.head(20))


cdystonia_long = cdystonia[['patient','site','id','treat','age','sex']].drop_duplicates().merge(
                    twstrs_wide, right_index=True, left_on='patient', how='inner')
print(cdystonia_long.head())


# To convert our "wide" format back to long, we can use the melt function, appropriately parameterized:
pd.melt(cdystonia_long, id_vars=['patient','site','id','treat','age','sex'], 
        var_name='obs', value_name='twstrs')
```
