```python
# use fisher exact test to see if hpv_16 is enriched in squamous or adeno
# contingency table:
#                               adeno   squamous
# number_of_hpv16+_patients:    1         39 
# number_of_hpv16-_patients:    18        24
import scipy.stats as stats
oddsratio, pvalue = stats.fisher_exact([[1, 39], [18, 24]])
print(pvalue)
```

```python
# hpv load
df = pd.read_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/NCI_HIV_Cervial_genome_and_transcriptome_BBT_results_03092016.csv', 
                 sep='\t', index_col='patient')

# look at genome phv load
df = df[df.data_type=='genome']
# df = df[df.data_type=='transcriptome']
df.columns
df = df[['HIV_status', 'Human_papillomavirus']]
# grouped = df.groupby('HIV_status')
df = df.sort_values(['HIV_status', 'Human_papillomavirus'], ascending=[True, False])
# colors = df_genome_sorted['HIV_status']
colors = {'Negative':'blue', 'Positive':'red'}
fig = plt.figure()
# all following 3 works, just pass a list of colors
# df_genome_sorted['genome_load'].plot(kind='bar', color=[colors[i] for i in df_genome_sorted['HIV_status']])
# df_genome_sorted['genome_load'].plot(kind='bar', color=list(df_genome_sorted['HIV_status'].map(colors)))
df['Human_papillomavirus'].plot(kind='bar', 
                                color=df['HIV_status'].apply(lambda x: colors[x]), 
                                title='Genome HPV load HIV + vs -')
plt.ylabel('Percent of total reads')

# draw customized legend
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

Negative = mpatches.Patch(color='blue', label='HIV Negative')
Positive = mpatches.Patch(color='red', label='HIV Positive')

plt.legend(handles=[Negative, Positive], loc=2)
plt.savefig('HPV load HIV + vs -', bbox_inches='tight')
```

```python
# get microbial load raw counts for sara's script'
df = pd.read_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/bbt_raw_counts_files.txt', 
                 sep='\t', header=None)
df.head()
bbt_files = df[0].tolist()
df_bbt = pd.DataFrame()
for f in bbt_files:
    patient = f.split('/')[5]
    df_tmp = pd.read_csv(f, sep='\t')
    df_tmp.set_index('filter_id', inplace=True)
    df_tmp = df_tmp[['hits']]

    df_tmp.head()

    df_tmp.rename(columns={'hits':patient}, inplace=True)
    df_tmp.head()

    if df_bbt.empty:
        df_bbt = df_tmp
    else:
        df_bbt = df_bbt.join(df_tmp)
df_bbt = df_bbt.T
df_bbt.to_csv('bbt_raw_counts.tmp', sep='\t')
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.stats import ttest_ind
import scipy.stats as ss
matplotlib.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15, 9)
# matplotlib.rcParams['figure.figsize'] = (3.75, 2.25)
import pandas as pd
import numpy as np
```

```python
# look at total microbial load
df_raw = pd.read_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients/BBT_matrix_no_normal.txt',
                 sep='\t', index_col='patient')

df_clinic = pd.read_excel('/projects/trans_scratch/validations/workspace/szong/Cervical/E6_E7_expression/Cervical_integration_expression_mastertable_reviewed.xls')
df_clinic.set_index('patient', inplace=True)
df_clinic.head(2)
# df_raw.columns

df_clinic = df_clinic[['putative_Histology', 'HIV_status']]
df_clinic.head(2)

df_clinic.shape

# look at genome microbial load
df_genome = df_raw[df_raw.data_type=='genome']
# including hpv as well 'Human_papillomavirus',
df_genome = df_genome.drop(['tissue_status', 'HIV_status', 'data_type', 'vectors', 'phiX', 'multiMatch', 'noMatch', 'human'], axis=1)
df_genome['genome_logRPM'] = np.log10(df_genome.sum(axis=1)*100*1000000+1)
df_genome = df_genome.loc[:, ['genome_logRPM']]
df_genome.head(2)

# look at transcriptome microbial load
df_transcriptome = df_raw[df_raw.data_type=='transcriptome']
# 'Human_papillomavirus',
df_transcriptome = df_transcriptome.drop(['tissue_status', 'HIV_status', 'data_type', 'vectors', 'phiX', 'multiMatch', 'noMatch', 'human'], axis=1)
df_transcriptome['transcriptome_logRPM'] = np.log10(df_transcriptome.sum(axis=1)*100*1000000+1)
df_transcriptome = df_transcriptome.loc[:, ['transcriptome_logRPM']]
df_transcriptome.head(2)

# # iloc support integer index
# # loc support label index
# # df.iloc[[2]]
# # df.loc[['HTMCP-03-06-02205']]
# # df.idxmax(axis=1)


df_joined = df_genome.join(df_transcriptome, how='outer')
df_joined = df_joined.join(df_clinic)

df_joined.head()
```

```python
df_joined.to_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients/microbial_load.txt', sep='\t')
```

```python
fig = plt.figure()
group_col = 'HIV_status'
# group_col = 'putative_Histology'
value_col = 'genome_logRPM'
ax = df_joined.boxplot(column=value_col, by=group_col, showfliers=False, widths=(0.6,0.6), figsize=(5,5))
plt.title('') # this removes the column as subtitle of the plot
fig = ax.get_figure()
# fig.suptitle('')

np.random.seed(0)
# overlay actuall data points
categories = df_joined[group_col].unique().tolist()
for category in categories:
    y = df_joined[value_col][df_joined[group_col]==category]
#     i = categories.index(category) + 1
    print category
    if category == 'Positive':
#     if category == 'Adeno':
        i = 2
    elif category == 'Negative':
#     elif category == 'Squamous':
        i = 1
    # add jitter and specify maker facecolor and edgecolor
    x = np.random.normal(i, 0.1, size=len(y))
    plt.plot(x, y, 'o', markerfacecolor='white',markeredgecolor='black', markersize=7, alpha=0.8)

# ax.set_ylim(0,0.1)
ax.set_axis_bgcolor('whitesmoke') #specify plot backgroup color
ax.tick_params(axis='x', labelsize=14) #specify x tick label size
plt.xlabel("HIV status", fontsize=14) 
plt.ylabel('genome_logRPM', fontsize=14)
plt.suptitle('genome microbial load HIV + vs -', fontsize=14)  # get rid of the automatically generated title
plt.savefig('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients/genome_microbial_load_101_patients.png', dpi=200)

# statistical test 
print('Mann Whitney U test on {0} comparing {1}'.format(value_col, group_col))
Negative = df_joined[df_joined[group_col]=='Negative']
# Negative = df_joined[df_joined[group_col]=='Squamous']
Positive = df_joined[df_joined[group_col]=='Positive']
# Positive = df_joined[df_joined[group_col]=='Adeno']
# tie corrected Mann Whitney U test,
ss.mannwhitneyu(Negative[value_col], Positive[value_col], use_continuity=True, alternative=None)
```

# process RNA bbt raw counts for sara's script to plot the microbial profile 

```python
import os
os.chdir('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients')

# get microbial load raw counts for sara's script'
df = pd.read_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients/101_rnaseq_bbt_rawcount_file.txt', 
                 sep='\t', header=None)
df.head()
bbt_files = df[0].tolist()
df_bbt = pd.DataFrame()
for f in bbt_files:
    splt = f.split('/')
    patient = splt[5]
    if len(splt) > 9:
        patient = splt[7]
#     patient = f.split('/')[5]
    df_tmp = pd.read_csv(f, sep='\t')
    df_tmp.set_index('filter_id', inplace=True)
    df_tmp = df_tmp[['hits']]

    df_tmp.head()

    df_tmp.rename(columns={'hits':patient}, inplace=True)
    df_tmp.head()

    if df_bbt.empty:
        df_bbt = df_tmp
    else:
        df_bbt = df_bbt.join(df_tmp)
df_bbt = df_bbt.T
df_bbt.to_csv('RNA_bbt_raw_counts.txt', sep='\t')

# add total to the file
ndf = pd.read_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients/RNA_bbt_raw_counts_formatted.txt', sep='\t')
ndf.head(2)

ndf['total'] =ndf.sum(axis=1)
ndf.head(2)

ndf.to_csv('/projects/trans_scratch/validations/workspace/szong/Cervical/bbt/101_patients/RNA_bbt_raw_counts_formatted.txt.new', sep='\t', index=False)
```
